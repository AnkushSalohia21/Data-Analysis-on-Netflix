```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Netflix.csv")
df
```

```python
df.isnull()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
df
```

```python
df['Subscription Type'].info()
```

```python
print(df['Subscription Type'].unique())
print(df['Country'].unique())
print(df['Device'].unique())



```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Subscription Type", data=df)
plt.title("Users by Subscription Type")
plt.xlabel("Subscription Type")
plt.ylabel("Number of Users")
plt.show()
```

```python
top_countries = df["Country"].value_counts().head(100)
plt.figure(figsize=(7,4))
top_countries.plot(kind="bar")
plt.title("Countries by Number of Users")
plt.xlabel("Country")
plt.ylabel("Users")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Subscription Type", hue="Gender", data=df)
plt.title("Gender vs Subscription Type")
plt.xlabel("Subscription Type")
plt.ylabel("Users")
plt.show()
```

```python
subscription_counts = df["Subscription Type"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(subscription_counts, labels=subscription_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Subscription Type Distribution")
plt.show()
```
